```python
import os

current_directory = os.getcwd()  # Get the current working directory
content_directory = os.path.join(current_directory, "content")  # Join the paths correctly

# Assuming SimpleDirectoryReader is correctly imported and available
documents = SimpleDirectoryReader(content_directory).load_data()  # Load data from the content directory
print(documents)
```

```python
system_prompt="""
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided.
"""
## Default format supportable by LLama2
query_wrapper_prompt=PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
import torch

llm = HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="chandankh/Llama-2-7b-chat-finetune-test",
    model_name="chandankh/Llama-2-7b-chat-finetune-test",
    device_map="auto",
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16 , "load_in_8bit":True}
)
```

```python
%pip install llama-index-embeddings-langchain
!pip install llama-index
```

```python
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from llama_index.core import ServiceContext
from llama_index.embeddings.langchain import LangchainEmbedding

embed_model=LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2"))
```

```python
service_context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

```python
service_context
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=service_context)
```

```python
index
```

```python
query_engine=index.as_query_engine()
```

```python
response=query_engine.query("Draw neat and labeled diagram of nephron and describe the process of urine formation. or else Describe structure and functioning of nephron")
```

```python
print(response)
```
